```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
  
def estimate_coef(x, y): 
    # number of observations/points 
    n = np.size(x) 
  
    # mean of x and y vector 
    m_x, m_y = np.mean(x), np.mean(y) 
  
    # calculating cross-deviation and deviation about x 
    SS_xy = np.sum(y*x) - n*m_y*m_x 
    SS_xx = np.sum(x*x) - n*m_x*m_x 
  
    # calculating regression coefficients 
    b_1 = SS_xy / SS_xx 
    b_0 = m_y - b_1*m_x 
  
    return(b_0, b_1) 
  
  
def plot_regression_line(x, y, b, label): 
    # plotting the actual points as scatter plot 
    plt.scatter(x, y, color = "m", 
               marker = "o", s = 30) 
  
    # predicted response vector 
    y_pred = b[0] + b[1]*x 
  
    # plotting the regression line 
    plt.plot(x, y_pred, color = "g", label=label) 
  
    # putting labels 
    plt.xlabel('Weight Data (lbs)') 
    plt.ylabel('Sensors SUM (Voltage SUM)') 
    plt.title('Lab Test in ginning lab')
    plt.legend(loc='best')
  
    # function to show plot 
    plt.show() 

# actual weight of seed cotton for Test 1 to 3 & Test 4 to 6
weight = np.array([5.0, 10.0, 15.0])

# Raw Sum Data of Sensor 1 & 2 
g_1 = np.array([551, 948.5, 1353])
g_2 = np.array([391, 717.8, 1117.8])

s1_1 = np.array([78233245.0, 112201991.0, 161403680.0])
s2_1 = np.array([78868730.0, 117642295.0, 195694506.0])

s1_2 = np.array([60995238.0, 109777576.0, 156733025.0])
s2_2 = np.array([86257589.0, 153166316.0, 214782246.0])

# Correct baseline with zcore algorithm
z1_1 = np.array([553839.28, 1363556.72, 1094709.87])
z2_1 = np.array([3505593.39, 7230431.12, 9994896.27])

z1_2 = np.array([539297.61, 669923.51, 816069.53])
z2_2 = np.array([3196333.49, 5816429.10, 8757617.80])

# Correct baseline with minimum value algorithm
m1_1 = np.array([863394.0, 1938731.0, 1880651.0])
m2_1 = np.array([8036181.0, 15957625.0, 25597183.0])

m1_2 = np.array([897238.0, 1045336.0, 1495675.0])
m2_2 = np.array([7493239.0, 14888996.0, 23562326.0])

# Convert DNs to Voltage 
coff = 32767*2
# Sensor 1
v1 = s1_1/coff
v2 = s1_2/coff

v3 = z1_1/coff
v4 = z1_2/coff

v5 = m1_1/coff
v6 = m1_2/coff


# Sensor 2
w1 = s2_1/coff
w2 = s2_2/coff

w3 = z2_1/coff
w4 = z2_2/coff

w5 = m2_1/coff
w6 = m2_2/coff

b1 =  estimate_coef(weight, v3)
label1= 'y = '+ str(round(b1[0],2))+' + ' + str(round(b1[1],2))+'*x'
plot_regression_line(weight, v3, b1, label1) 
#from sklearn.metrics import r2_score
p1 = [b1[0]+(b1[1]*x) for x in weight]
print("Mean absolute error: %.2f"
      % mean_absolute_error(v3, p1)) 
print("Mean absolute percentage error: %.2f"
      % mean_absolute_percentage_error(v3, p1)) 
print("R squared value: %.2f" % r2_score(v3, p1))

b2 =  estimate_coef(weight, v4)
label2= 'y = '+ str(round(b2[0],2))+' + ' + str(round(b2[1],2))+'*x'
plot_regression_line(weight, v4, b2, label2) 
#from sklearn.metrics import r2_score
p2 = [b2[0]+(b2[1]*x) for x in weight]
print("Mean absolute error: %.2f"
      % mean_absolute_error(v4, p2))
print("Mean absolute percentage error: %.2f"
      % mean_absolute_percentage_error(v4, p2)) 
print("R squared value: %.2f" % r2_score(v4, p2))

# combine test 1 to 6
c1 = np.concatenate((weight,weight))
c2 = np.concatenate((v3, v4))

b3 =  estimate_coef(c1, c2)
label3= 'y = '+ str(round(b3[0],2))+' + ' + str(round(b3[1],2))+'*x'
plot_regression_line(c1, c2, b3, label3) 
#from sklearn.metrics import r2_score
p3 = [b3[0]+(b3[1]*x) for x in c1]
print("Mean absolute error: %.2f"
      % mean_absolute_error(c2, p3)) 
print("Mean absolute percentage error: %.2f"
      % mean_absolute_percentage_error(c2, p3)) 
print("R squared value: %.2f" % r2_score(c2, p3))





################ 0408 version #############

    

```
